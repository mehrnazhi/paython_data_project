```python
class BaseSalary:
    def __init__(self, salary_base, bonus_base = 0.1, symbol = '$'):
        self.salary_base = salary_base
        self.bonus_base = bonus_base
        self.symbol = symbol
        self.total_salary = salary_base * (1+ bonus_base)
        self.bonus = self.total_salary - self.salary_base

    def __repr__(self):
        return f'{self.salary_base:,.0f}{self.symbol}'

    def show_salary(self):
        return f'{self.total_salary:,.0f}{self.symbol}'
    def show_bonus(self):
        return  f'{self.bonus:,.0f}{self.symbol}'
```

```python
salary = BaseSalary(100000)
salary 
```

```python
salary.show_bonus
salary
```
